Em **python**, podemos também definir uma variável do tipo booleana, a qual pode assumir valores ```True``` (Verdadeiro) ou ```False``` (Falso). Além disso, as expressões podem ser avaliadas como ```True``` ou ```False```. O operador lógico **NÃO (not)** serve para checar se uma expressão é falsa. Seja:

\\





```
not(a == b)
```

\\

A expressão acima será verdadeira se ```a```, de fato, não for igual a ```b```. Ou seja, NÃO FALSO = VERDADEIRO.

\\



````python
x = False
if (x):
  print("Verdadeiro!")
else:
  print("Falso!")
````

````python
a = 107
b = 107
if ( not (a == b) ):
  print("De fato, a e b são DIFERENTES!")
else:
  print("De fato, a e b são IGUAIS!") 
  
````
