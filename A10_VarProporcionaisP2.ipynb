## **Variáveis Proporcionais: Razões e Proporções**

---

A **razão** entre dois números $a$ e $b$, $b \not = 0$, é:

$$
\frac{a}{b}
$$


\\

Quando duas ou mais razões possuem o mesmo resultado, elas são denominadas de **proporção**. Assim, a proporção está condicionada à razão entre a e b que é igual à razão entre c e d, e podemos exprimir assim:

$$
\frac{a}{b}=\frac{c}{d}, \quad \mbox{onde } b \not = 0, d \not = 0 
$$



**Exercício Resolvido 1**: Divida o número 2.500 entre duas partes diretamente proporcionais a 3 e 7.

\\

*Resp:* Nesse caso, a proporção é a seguinte:
$$
\frac{x}{3}=\frac{y}{7} 
$$

\\

O valor proporcional, $p$, é calculado assim:

$$
p = \frac{2500}{3 + 7} = 250
$$

\\

Então:

$$
\frac{x}{3} =  250 \Longrightarrow x = 750
$$

$$
\frac{y}{7} =  250 \Longrightarrow y = 1750
$$

\\

Logo, $x$ e $y$ podem ser calculados assim:

$$
x = p \cdot parte_x \quad \mbox{e} \quad y = p \cdot parte_y
$$


```
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from fractions import Fraction
```

```
def calcular_proporcional(x1, y1, x2, t):
  if (t == 'd'): # diretamente proporcional
    y2 = (x2*y1) / x1
  elif (t == 'i'): # inversamente proporcional
    y2 = (x1*y1) / x2
  else:
    print('Opção Inválida!')
  return Fraction(y2).limit_denominator()    
```

```
def dividir_proporcao(parte_x, parte_y, valor):
  p = (valor) / (parte_x + parte_y)
  # x/parte_x = prop
  x = p * parte_x
  # y/parte_y = prop
  y = p * parte_y
  return Fraction(x).limit_denominator(), Fraction(y).limit_denominator()
```

```
def calc_aum_dim(i, f, t):
  if (t == 'a'): # aumento
    if (f < i):
      print('Erro: valor final tem que sair maior ou igual ao valor inicial!')
    else:  
      perc = ((f - i) / i) * 100
      print('O aumento percentual foi de: ', perc , '%' )
  elif (t == 'd'): # diminuição
    if (i < f):
      print('Erro: valor inicial tem que sair maior ou igual ao valor fim!')
    else: 
      perc = ((i - f) / i) * 100
      print('A diminuição percentual foi de: ', perc, '%')
  else:
    print('Opção Inválida!')
  
```

```
x, y = dividir_proporcao(3, 7, 2500)
print('O valor de x é: ', x, ' e o valor de y  é: ', y)
```

## **Variáveis Proporcionais: Porcentagens**

---

A **porcentagem** de um número é:

$$
x \% \mbox{ de } y = \frac{x}{100} \cdot y 
$$


\\

A porcentagem de um número em relação a outro número é:

$$
\frac{x}{y} \cdot 100\% 
$$


\\

O aumento percentual é dado por:

$$
\frac{x_{final} - x_{inicial}}{x_{inicial}} \cdot 100\%
$$

\\

A diminuição percentual é dada por:

$$
\frac{x_{inicial} - x_{final}}{x_{inicial}} \cdot 100\%
$$


**Exercício Resolvido 2**: Retirando 50 de 380, qual foi a diminuição percentual?

\\

*Resp:* 
$$
dim = \frac{380 - 330}{380} = \mbox{13,16} \% 
$$



```
calc_aum_dim(380, 330, 'd')
```

**Exercício Resolvido 3**: Uma concessionária vende um carro à vista por R\$ $20.000\mbox{,00}$.
O mesmo carro pode ser vendido a prazo em até 36 parcelas de R\$ $600\mbox{,00}$. Qual o valor da taxa de juros usada? 

\\

*Resp:* Nesse caso, o valor final é de $36 \cdot 600 = 21600$. A taxa de juros é um tipo de aumento percentual ($aum$). Logo:

$$
aum = \frac{21600 - 20000}{20000} = 8 \% 
$$


```
calc_aum_dim(20000, 21600, 'a')
```
