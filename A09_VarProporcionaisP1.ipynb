## **Variáveis Proporcionais**

---

Duas variáveis $x$ e $y$, num certo universo, são **diretamente proporcionais** se o seguinte é válido, para $k \in \mathbb{R_{+}^{*}}$:

$$
y = kx
$$

\\

Nesse caso, dizemos que $y$ é diretamente proporcional a $x$.

\\

Em algumas situações, não se faz necessário saber o valor de $k$, conhecida como constante de proporcionalidade, para resolver certos problemas. Nesse caso, pode-se fazer:

$$
\frac{x_1}{x_2}=\frac{y_1}{y_2}
$$

\\

Assumindo que sempre se deseja saber o valor $y_2$ dados os outros três valores das variáveis, então:

$$
y_2 = \frac{x_2  y_1}{x_1}
$$


**Exercício Resolvido 1**: Se $x$ e $y$ são diretamente proporcionais, e se quando $x = 8$ temos $y = 3$, qual o valor de $y$ quando $x = 14$?

\\

*Resp:* 
$$
\frac{8}{14}=\frac{3}{y} \Longrightarrow y = \frac{3 \cdot 14}{8} \Longrightarrow y = \frac{21}{4}
$$

````
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from fractions import Fraction
````

````
def calcular_proporcional(x1, y1, x2, t):
  if (t == 'd'): # diretamente proporcional
    y2 = (x2*y1) / x1
  elif (t == 'i'): # inversamente proporcional
    y2 = (x1*y1) / x2
  else:
    print('Opção Inválida!')
  #return y2
  return Fraction(y2).limit_denominator()    
````

````
y = calcular_proporcional(8, 3, 14, 'd')
print('Valor de y: ', y)
````

````
x = np.arange(0, 16)
print("Esses são os valores de x: ", x)
for k in range(1,17,5):
  y = []
  y = k*x
  if (k == 1):
    cor = 'b'
  elif (k == 6):
    cor = 'r'
  elif (k == 11):
    cor = 'g'
  else: # magenta
    cor = 'm'      
  plt.plot(x, y, cor, label = 'k = '+str(k))
  plt.axis([0, max(x)+2, 0, max(y)+7])
  plt.xticks(np.arange(0, max(x)+2, step=1.0))
  plt.legend()
    

plt.grid()
plt.title("Diretamente Proporcionais")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````


\\

Duas variáveis $x$ e $y$, num certo universo, são **inversamente proporcionais** se o seguinte é válido, para $k \in \mathbb{R_{+}^{*}}$:

$$
y = \frac{k}{x}
$$

\\

Nesse caso, dizemos que $y$ é inversamente proporcional a $x$. Assim como no caso anterior, em algumas situações, não se faz necessário saber o valor de $k$ para resolver alguns problemas. Podemos fazer o uso de:

$$
x_1 y_1 = x_2 y_2
$$

\\

Assumindo que sempre se deseja saber o valor $y_2$ dados os outros três valores das variáveis, então:

$$
y_2 = \frac{x_1 y_1}{x_2}
$$







**Exercício Resolvido 2**: Se $x$ e $y$ são inversamente proporcionais, e se quando $x = 13$ temos $y = 2$, qual o valor de $y$ quando $x = 24$?

\\

*Resp:* 
$$
13 \cdot 2 = 24 \cdot y \Longrightarrow y = \frac{13 \cdot 2}{24} \Longrightarrow y = \frac{13}{12}
$$

````
y = calcular_proporcional(13, 2, 24, 'i')
print('Valor de y: ', y)
````

````
x = np.arange(1, 16)
print("Esses são os valores de x: ", x)
for k in range(1,17,5):
  y = []
  y = k/x
  if (k == 1):
    cor = 'b'
  elif (k == 6):
    cor = 'r'
  elif (k == 11):
    cor = 'g'
  else: # magenta
    cor = 'm'      
  plt.plot(x, y, cor, label = 'k = '+str(k))
  plt.axis([0, max(x)+2, 0, max(y)+1])
  plt.xticks(np.arange(0, max(x)+2, step=1.0))
  plt.legend()
    

plt.grid()
plt.title("Inversamente Proporcionais")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````
