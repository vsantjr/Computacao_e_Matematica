## **Quadriláteros**
----

Polígonos que possuem quatro lados, e podem ser classificados como:

\\

*Trapézio*: Quadrilátero que possui dois lados paralelos. Existem três tipos de trapézio:



* Trapézio Retângulo: Trapézio que apresenta dois ângulos retos;
* Trapézio Isósceles: Trapézio que apresenta dois lados congruentes (possuem a mesma medida) e dois lados diferentes;
* Trapézio Escaleno: Trapézio em que todos os lados apresentam medidas diferentes.



\\

*Paralelogramo*: Quadrilátero que possui dois pares de lados paralelos.




\\

*Losango*: Quadrilátero que possui todos os lados congruentes.

\\

*Retângulo*: Quadrilátero que possui todos os ângulos internos congruentes.







\\

*Quadrado*: Quadrilátero que possui todos os lados e ângulos internos congruentes.

\\

**Área de Quadriláteros**



*Quadrado e Retângulo*

Para calcular a área ($A$) de um quadrado ou retângulo, usamos a fórmula:
$$ A = base \times altura $$

\\

*Paralelogramo*

A área ($A$) de um paralelogrma também é calculada pela fórmula:
$$ A = base \times altura $$


Mas, nesse caso, é preciso determinar a altura pelo Teorema de Pitágoras. A base, usualmente, é considerada como o lado inferior.


<img src="https://drive.google.com/uc?id=172g7CIZKIX5Su2FG-fOCZ88DmujoYYLm" alt="Drawing" width="350"/>


\\

*Losango* 

Losangos, quadrados e retângulos são todos paralelogramos, mas nem todo paralelogramo é um losango ou quadrado ou retângulo. Então, a área ($A$) de um losango também pode ser calculada pela fórmula abaixo, com o auxílio do Teorema de Pitágoras para o cálculo da altura:

$$ A = base \times altura $$

Uma maneira alternativa para calcular a área de um losango é dada pela fórmula a seguir:



$$ A = \frac{d \times D}{2} $$

onde $d$ e $D$ são as diagonais menores e maiores do losango, respectivamente.

<img src="https://drive.google.com/uc?id=1dPe92cEFvqn452URrc_0DhymMO-OIgXt" alt="Drawing" width="350"/>



*Trapézio*

A área ($A$) do trapézio pode ser calculada pela fórmula:

$$
A = \frac{(base\_menor + base\_maior) \times altura}{2}
$$

```
import numpy as np
import matplotlib.pyplot as plt
import math
from fractions import Fraction
```

```
def calcular_area_quadrilatero(t, b, h, bM):
  area = -1
  # Quadrado, Retângulo, Paralelogramo: b = base (lado); h = altura
  if (t == 'q') or (t == 'r') or (t == 'p'): 
    area = b*h
  # Losango: b = d; h = D
  elif (t == 'l'): 
    if (b >= h):
      print("Diagonal menor não é de fato menor do que a diagonal maior!")
    else:  
      area = (b*h)/2 
  # Trapézio: b = base menor; bM = base maior; h = altura
  else: # Trapézio
    if (b >= bM):
       print("Base menor não é de fato menor do que a base maior!")
    else:   
       area = ((b + bM)*h)/2   
  return area     

        
```

```
cond = False

while not cond:
  try: 
    tipo = input("Entre com o tipo de quadrilátero (q, r, p, l, t): ")
    base = float(input("Entre com a base/diagonal menor/base menor: "))
    if (tipo == 'q'):
      altura = base
    else:  
      altura = float(input("Entre com a altura/diagonal maior: "))
    if (tipo == 't'):
      base_maior = float(input("Entre com a base maior: "))
    else:
      base_maior = 1
    cond = True
  except ValueError:
    print("Entre apenas com números!")
  
# Os lados devem ser maior do que Zero
if base<=0 or altura<=0 or base_maior<=0:
  print("As medidas devem ser maior do que Zero!")  
else:
  print("\n----------")
  ar = calcular_area_quadrilatero(tipo, base, altura, base_maior)
  print("\nÁrea: ", ar)

```
