**Gráficos**

Visualizar as informações, funções, resultados é algo muito relevante para o entendimento de uma disciplina como Matemática. A biblioteca *Matplotlib* do **python** é outra muito usada pelos programadores, e que serve para criar/mostrar gráficos estáticos, interativos e animados. Fazer um gráfico é popularmente conhecido como *plotar*.

\\

Para importar a biblioteca *Matplotlib* podemos fazer:

```
import matplotlib.pyplot as plt
```




````
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
````

````
def funcao_grafico(x):
  c = 4.0/(np.pi) 
  y = np.zeros(len(x))
  for n in range (1,1001,2):
    y = y + ((1/np.pi) * np.sin(n*np.pi*x))
  
  y = c*y*np.exp(-x)
  return y  
      
````

````
x = [1, 2, 3, 4, 5]
y = [1, 3, 9, 10, 7]
plt.plot(x, y, 'bv--')
plt.axis([0, 6, 0, 11])
plt.xticks(np.arange(0, 6, step=1.0))
plt.yticks(np.arange(0, 11, step=1.0))
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Primeiro Gráfico")
plt.show()
````

````
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]
plt.plot(x, y1, 'b', label="Azul")
plt.plot(x, y2, 'g', label="Verde")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Dois Gráficos")
plt.legend()
plt.show()
````

````
t1 = np.arange(-np.pi, np.pi, 0.1)
t2 = np.arange(0.0, 4.0, 0.02)

plt.figure()
plt.subplot(221)
plt.title('Gráfico 1')
plt.plot(t1, funcao_grafico(t1), 'm')

plt.subplot(222)
plt.title('Gráfico 2')
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.subplot(223)
plt.title('Gráfico 3', y=-0.4)
plt.plot(t2, np.sin(2*np.pi*(t2**2)), 'g')

plt.subplot(224)
classes = ['5A', '6A', '7A', '8A', '9A']
pontos = [10, 25, 70, 88, 80]
plt.title('Gráfico 4', y=-0.4)
plt.bar(classes, pontos)
#plt.show()
plt.savefig('varios.png',bbox_inches='tight')
files.download("varios.png")
````

````
fig = plt.figure()
ax = fig.gca(projection='3d')

# Dados
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Mostra o gráfico
surf = ax.plot_surface(X, Y, Z, cmap=cm.RdYlGn,
                       linewidth=0, antialiased=False)

# Ajusta o eixo z
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Adiciona uma barra de cor
fig.colorbar(surf, shrink=0.5, aspect=7)
plt.show()
````
