## **Estruturas de Controle (Decisão)**

As estruturas de controle (decisão) permitem alterar o curso do fluxo de execução de um programa. Isso depende do resultado de um teste lógico que pode ser:


*   *True* (Verdadeiro);
*   *False* (Falso).

\\

<img src="https://drive.google.com/uc?id=1BbQ3UDqSQY0IuQg3O37TeIa7VAVJxidv"
alt="Drawing" width="575"/>

Image by <a href="https://pt.dreamstime.com/photos-images/bifurca%C3%A7%C3%A3o.html">Dreamstime</a>


\\

Em python, existem, basicamente, essas estruturas de controle:



1.   ***if*** $\Longrightarrow$ se
2.   ***if...else*** $\Longrightarrow$ se...senão
3.   ***if...elif...else*** $\Longrightarrow$ se...senão se...senão




```
a = 190
if (a > 100):
  print("Maior ou igual a 100!")
  print("mais um")
else:
  print("Menor do que 100!") 
  print("outro") 

print("continuando!")
```

```
a = 85
if (a < 50):
  print("Menor do que 50!")
elif (a > 100):
  print("Maior do que 100!")
else:
  print("Maior do que 50 e menor do que 100!")  

print("continuando!")
```
