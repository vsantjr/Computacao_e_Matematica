**Estruturas de Repetição**

As estruturas de repetição permitem que um bloco de comandos seja executado
diversas vezes. Também são popularmente conhecidas como laços (*loops*). 

\\

<img src ="https://drive.google.com/uc?id=1e7D_D-proBiAgiuzpNYb0_hIRJ4MBmij"
alt="Drawing" width="300"/>

Image by <a href="https://www.pngwing.com/pt/free-png-zdyim">PNGWING.</a>

\\

Existem dois tipos de repeticão:

*   repetição contável: executa um bloco de comandos por um número predeterminado de vezes. Comando: ```for```;
*   repetição condicional: executa um bloco de código
enquanto uma condição lógica for verdadeira. Comando: ```while```.





**Comando For**

O comando ```for``` pode ser usado com uma sequência numérica (gerada com o comando ```range```) ou associado a uma lista. O trecho de código da
repetição é executado para cada valor da sequência numérica ou da lista.


\\
*Importante 1:* Nas opções abaixo quando aparece o número que indica o **fim**, a função ```range``` faz com que o laço termine sempre no elemento imediatamentte anterior a **fim**. 

\\



Sintaxe - opção 1:

\\

```for``` variável ```in range``` (início, fim, passo):

&emsp; {bloco de comandos}

\\

Sintaxe - opção 2:

\\

```for``` variável ```in``` lista:

&emsp; {bloco de comandos}

\\

*Importante 2:* Existem comandos adicionais que podem ser usados tanto no comando ```for``` como no comando ```while``` (vide adiante) para melhor controlar a execução dos laços. Tais comandos são:



1.   ```continue```: serve para pular o restante do código dentro de um laço, mas apenas para a iteração atual. O laço não termina, mas continua com a próxima iteração;
2.   ```break```: finaliza o laço. O fluxo do programa continua no comando imediatamente após o corpo do laço;
3.   ```else```: executado quando o laço termina por exaustão da variável de iteração (com ```for```) ou quando a condição se torna falsa (com ```while```), mas não quando o laço é finalizado por uma instrução ```break```.






````
import numpy as np
````

````
x = 1
print('Valor de x: ', x)
x+=1     # Atenção: x+=1 é o mesmo que x = x + 1
print('Valor de x: ', x)
x+=1
print('Valor de x: ', x)
x+=1
print('Valor de x: ', x)
x+=1
print('Valor de x: ', x)

# Com o comando for: sintaxe 1 (passo omitido: valor padrão igual 1)
print('\n---------------\n')
for x in range(1,101):
  print('Valor de x: ', x)
````

````
# Comando for: sintaxe 1. Passo igual a 3.
for x in range(20,31,3):
  print('Valor de x: ', x)
````

````
# Comando for: sintaxe 2.
lst = np.arange(20,31,3)
for x in lst:
  print('Valor de x: ', x)
````

````
# input: modo de solicitar entrada via teclado.
palavra = input('Palavra: ')
for x in palavra:  
  if ( (x == 'a') or (x == 'e') ): 
    continue 
  print('Não é "a" e nem "e": ', x)         

print('continuando!')
````

**Comando While**

O comando ```while``` executa o bloco de comandos enquanto uma condição
for verdadeira. A condição é uma expressão booleana que pode fazer
uso de quaisquer operadores.

\\


Sintaxe:

\\

```while``` condição:

&emsp; {bloco de comandos}








````
soma_total = 0
x = int(input('Número: '))
while x != 0:
  soma_total += x     # O mesmo que: soma_total = soma_total + x
  if soma_total > 20:
    print('A soma total é: ', soma_total)
    break           
  x = int(input('Outro No.: '))

print('continuando!')
````
