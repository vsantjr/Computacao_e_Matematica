## **Ângulo Inscrito Particular**

---

Uma reta, ou semirreta ou segmento de reta, e uma circunferência são **tangentes** quando possuem somente um ponto comum, conhecido como **ponto de tangência**. Na figura abaixo, o raio $\overline{OB}$ é perpendicular à reta tangente.

Um outro conceito importante relacionado à circunferência é a **corda**. A corda é qualquer segmento de reta que liga dois pontos da circunferência. Na figura abaixo, $\overline{BC}$ é uma corda. Além disso, percebamos a semirreta $\overrightarrow{BA}$, a qual é tangente a circunferência no ponto $B$.


\\

\\

Então, o ângulo particular $A\hat{B}C$ é inscrito na circunferência e "enxerga" o arco $\overset{\Large\frown}{BC}$. Logo, podemos também assinalar que: $m_a(\overset{\Large\frown}{BC}) = m(B\hat{O}C) = 2 \times m(A\hat{B}C)$.  



## **Função de Cálculo de Medidas Angulares**

Para a função de ```calcular_medidas_angulares``` a seguir, vamos considerar a seguinte figura.

````
import numpy as np
import matplotlib.pyplot as plt
import math
from fractions import Fraction
import sys
````

````
def calcular_medidas_angulares(angulo_dado, valor_angulo_dado, angulo_resp):
  # angulo_dado: tipo do ângulo dado como entrada (x, y, z, i, w)
  # valor_angulo_dado: valor do ângulo dado como entrada
  # angulo_resp: tipo do ângulo que se quer obter a resposta da questão
  
  # valor_angulo_resp: valor do ângulo que se quer obter a resposta da questão.
  # Valor inicial = +infinito
  valor_angulo_resp = sys.float_info.max

  if ( ( (angulo_dado == 'x') or (angulo_dado == 'w') ) and (angulo_resp == 'i') ):
    valor_angulo_resp = valor_angulo_dado/2

  elif ( ( (angulo_dado == 'x') and (angulo_resp == 'w') ) or  
        ( (angulo_dado == 'w') and (angulo_resp == 'x') ) ):
    valor_angulo_resp = valor_angulo_dado

  elif ( ( (angulo_dado == 'y') or (angulo_dado == 'z') ) and (angulo_resp == 'i') ):
    x = 180 - 2*valor_angulo_dado 
    valor_angulo_resp = x/2

  elif ( ( (angulo_dado == 'y') or (angulo_dado == 'z') ) and (angulo_resp == 'w') ):   
    w = 180 - 2*valor_angulo_dado 
    valor_angulo_resp = w

  elif ( (angulo_dado == 'i') and ( (angulo_resp == 'x') or (angulo_resp == 'w') ) ): 
    valor_angulo_resp = 2*valor_angulo_dado

  else:
    print('Opção Inválida!')  

  return Fraction(valor_angulo_resp).limit_denominator()
  
````

````
med_angular = calcular_medidas_angulares('i', 68, 'x')
print('Valor da medida angular: {} graus'.format(med_angular))
print('#'*40+'\n')
````

## **Teorema das Cordas**
----

Se uma reta passa pelo ponto $P$ e corta uma circunferência nos
pontos $A$ e $B$, então o produto $PA \times PB$ é constante.

\\

### **Caso 2: Ponto $P$ é exterior à circunferência**

\\

<img src ="https://drive.google.com/uc?id=1Az3md8GTJzZtSL_tKtuvXAEvr29ADGU_"
alt="Drawing" width="300"/>



Image by <a href="http://moodle.profmat-sbm.org.br/MA13/Unidades/unidade11.pdf">Eduardo Wagner, SBM.</a>

\\

Os triângulos PAD e PCB são semelhantes. Desse modo, $\frac{PA}{PC} = \frac{PD}{PB}$. Portanto:

\\

$$
PA \times PB = PC \times PD = \mbox{constante}
$$

\\

````
def calcular_comprimento_cordas(pa, pb, pc, opcao):
  pd = (pa*pb)/pc
  comprimento = sys.float_info.max

  if (opcao == 'AB'):
    comprimento = pa+pb
  elif (opcao == 'CD'):
    comprimento = pc+pd
  elif (opcao == 'soma'):
    comprimento = pa+pb+pc+pd
  else:
    print('Opção Inválida!')  

  return Fraction(comprimento).limit_denominator()
````

````
comp_cordas = calcular_comprimento_cordas(2, 12, 10, 'AB')
print('Valor do comprimento das cordas: {}'.format(comp_cordas))
print('#'*40+'\n')
````

````
comp_cordas = calcular_comprimento_cordas(2, 12, 10, 'CD')
print('Valor do comprimento das cordas: {}'.format(comp_cordas))
print('#'*40+'\n')
````

````
comp_cordas = calcular_comprimento_cordas(2, 12, 10, 'soma')
print('Valor do comprimento das cordas: {}'.format(comp_cordas))
print('#'*40+'\n')
````

## **Cordas: Ângulos e Arcos**
----

Seja a figura abaixo. 

\\

<img src ="https://drive.google.com/uc?id=1wUsF8TeZQc7v1QTyzvSnX9C13oECjtsO"
alt="Drawing" width="300"/>

Image by <a href="http://objetoseducacionais2.mec.gov.br/bitstream/handle/mec/10396/geo0204.htm">MEC.</a>

\\

Considerando o triângulo $ABC$, os ângulos internos $b$ e $c$ são inscritos à circunferência. Desse modo, 

\\

$$
b = \frac{m_a(\overset{\Large\frown}{EC})}{2} = \frac{L'}{2}
$$ 



$$ 
c = \frac{m_a(\overset{\Large\frown}{BD})}{2} = \frac{L''}{2}
$$

\\

Mas, o ângulo $a$ é externo ao triângulo ABC. Logo:

$$
a = b + c \implies a = \frac{L'}{2} + \frac{L''}{2} \implies a = \frac{L' + L''}{2}
$$

\\

Obs: Um **ângulo excêntrico interior** de uma circunferência é todo ângulo definido por duas cordas que se cruzam em um ponto interior da circunferência, diferente do centro. O ângulo $a$ na figura acima é um ângulo excêntrico interior.



````
def calcular_medidas_angulares_2(valor_angulo_1, valor_angulo_2, angulo_resp):
  # valor_angulo_resp: valor do ângulo que se quer obter a resposta da questão.
  # Valor inicial = +infinito
  valor_angulo_resp = sys.float_info.max

  '''
  Se a medida angular que se quer é o a (excêntrico interior):
  valor_angulo_1 = L' (EC); valor_angulo_2 = L'' (BD)

  Se a medida angular que se quer é de qualquer um dos arcos: 
  valor_angulo_1 = a; valor_angulo_2 = L' (EC) ou L'' (BD)
  '''
  
  if (angulo_resp == 'a'):
    valor_angulo_resp = (valor_angulo_1 + valor_angulo_2)/2
  
  elif ( (angulo_resp == 'EC') or (angulo_resp == 'BD') ):
    valor_angulo_resp = 2*valor_angulo_1 - valor_angulo_2

  else:
    print('Opção Inválida!')  

  return Fraction(valor_angulo_resp).limit_denominator()
  
````

````
med_angular2 = calcular_medidas_angulares_2(40, 70, 'a')
print('Valor da medida angular: {} graus'.format(med_angular2))
print('#'*40+'\n')
````

````
med_angular2 = calcular_medidas_angulares_2(40, 70, 'BD')
print('Valor da medida angular: {} graus'.format(med_angular2))
print('#'*40+'\n')
````

````
med_angular2 = calcular_medidas_angulares_2(40, 70, 'EC')
print('Valor da medida angular: {} graus'.format(med_angular2))
print('#'*40+'\n')
````
