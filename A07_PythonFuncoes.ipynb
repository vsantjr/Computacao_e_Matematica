**Funções**

Assim como **python** tem uma série de funções disponíveis em bibliotecas, você pode criar suas próprias funções para serem reusadas. Então, funções são partes de código que são encapsuladas e que podem ser chamadas de qualquer outra parte do programa. 

\\

Uma função é definida assim:

\\

```def``` nome_função(parâmetros):

&emsp; {bloco de comandos da função}

&emsp; ```return``` {valores}

\\

*Importante 1*: Parâmetros são as variáveis que podem existir na interface de uma função, e eles são colocados entre os parênteses da mesma. Quando a função é chamada em outra parte do programa, são passados valores para tais variáveis. Esses valores são denominados argumentos. Mas, **uma função pode ou não ter parâmetros.**

\\

*Importante 2*: Uma função pode retornar um ou mais valores por meio do comando ```return```. Mas, **esse comando também é opcional** e, caso o mesmo não seja usado, então a função retorna o valor *None* (nenhum valor).

````
import math
````

````
def simples():
  print('Esse é um exemplo de função sem parâmetros e sem retorno de valores!')
  
````

````
def multiplicacao(a, b):
  print('Esse é um exemplo de função com parâmetros e sem retorno de valores!')
  print('Resultado da multiplicação: ', a*b)
  
````

````
def raiz_segundo(a, b, c):
  print('Esse é um exemplo de função com parâmetros e com retorno de valores!')
  x1 = x2 = math.inf    # Número infinito
  delta = b**2 - 4*a*c
  if (delta < 0):
    # Nesse caso, os valores infinitos representam que não tem raízes reais
    print('Delta é Negativo!')
  elif (delta == 0):
    print('Delta é Zero!')  
    x1 = x2 = -b / (2*a)
  else:  
    print('Delta é Positivo!') 
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    
  return x1, x2
  
````

````
simples()
````

````
multiplicacao(2,7)
````

````
r1 , r2 = raiz_segundo(1,-5,6)
print('As raízes da equação do 2o. grau são: ', r1, '  e  ', r2)
````

````
multiplicacao(6,40)
````

````
r1 , r2 = raiz_segundo(1,4,-10)
print('As raízes da equação do 2o. grau são: ', r1, '  e  ', r2)
````
