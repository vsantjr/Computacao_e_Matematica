**Listas**

Uma Lista é um conjunto sequencial de valores, onde cada valor é identificado via um índice.

\\

<img src ="https://drive.google.com/uc?id=1iX4eaXpWloLK8o08DyDEzPFxBXi_Lam2"
alt="Drawing" width="600"/>


````
import numpy as np
from math import sqrt
````

````
lst = [10, 11, 12, 13, 14]
print('Lista: ', lst)
````

**Acessar/Alterar Elementos de uma Lista**

Para acessar e alterar um elemento de uma lista, você usa o seu respectivo índice.

\\

Uma lista pode ter valores de diferentes tipos de variáveis, e inclusive ter uma outra lista como um de seus elementos.





````
print('1o. elemento: ', lst[0])
print('2o. elemento: ', lst[1])
print('3o. elemento: ', lst[2])
print('4o. elemento: ', lst[3])
print('5o. elemento: ', lst[4])

# Novo valor do 4o. elemento da lista

print('\n-----------------------------\n')
print('Novo valor: 4o. elemento')
lst[3] = 500
print('1o. elemento: ', lst[0])
print('2o. elemento: ', lst[1])
print('3o. elemento: ', lst[2])
print('4o. elemento: ', lst[3])
print('5o. elemento: ', lst[4])
````

````
lst_diferente = [3, 8.98, 'blz', [7, 8, 9]]
print('Lista diferente: ', lst_diferente)
print('Lista diferente - 3o. elemento: ', lst_diferente[2])
print('Lista dentro da lista: ', lst_diferente[3][1])
````

**Python** possui inúmeras bibliotecas já prontas, com código já criado, de forma que basta você saber usar as funções/métodos que estão nessas bibliotecas. Ou seja, você não precisa sair escrevendo um monte de código, podendo reusar o que já está disponível.

Para usar uma biblioteca, você deve utilizar o comando ```import```, como mostramos no início do programa:


\\

``` import numpy as np ```

\\

O que estamos fazendo aqui é usar a biblioteca *Numpy* do **python**. A *Numpy* é uma biblioteca **python** muito adequada para computação científica e, portanto, para o uso da Matemática.

\\

A instrução ```import``` diz isso ao **python**: importe a biblioteca *Numpy* como ```np```. Então, você pode escrever uma instrução ```np.nome_de_função``` para usar uma função da biblioteca *Numpy*. 

\\

Se você quiser importar uma função específica de uma biblioteca, você pode usar o comando ```from```:

``` from math import sqrt ```






````
# Segunda forma
lst2 = np.arange(10,15)
print('Lista - Segunda: ', lst2)

# Raiz quadrada
x = 16
print("A raiz quadrada de x é: ", sqrt(x))
````
