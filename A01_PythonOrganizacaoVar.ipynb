````
# Importação de bibliotecas, funções do python
import numpy as np


# Definição de suas funções
def minha_funcao(s):
  return s**2


# Escrita do corpo principal do seu programa
x = 10*np.pi
y = minha_funcao(x)
print('O resultado é: ', y)
````

## **Variáveis**


Variáveis são regiões de memória do computador, utilizadas para armazenar e manipular dados.


<img src="https://drive.google.com/uc?id=1P_0bNDELdQQ0whERN4dY5q0e0VUd63eN"
alt="Drawing" width="600"/>

\\

 Existem três tipos básicos de variáveis:


*   inteiro (armazena números inteiros; equivalente ao conjunto de números inteiros $\mathbb{Z}$);
*   float (armazena números em formato decimal; equivalente ao conjunto de números reais $\mathbb{R}$);
*   string (armazena um conjunto de caracteres).
 



````
x = 73
y = x
x = x + 2
print('O valor de x é: ', x)
print('O valor de y é: ', y)
print('\n----------\n')
print('O tipo da variável x é: ', type(x))
print('O tipo da variável y é: ', type(y))
````

````
x = 10.54
print('O valor de x é: ', x)
print('O tipo da variável x é: ', type(x))
````

````
x = 'Eu gosto de Tik Tok e WhatsApp! Tudo bem????'
print('O valor de x é: ', x)
print('O tipo da variável x é: ', type(x))
````
