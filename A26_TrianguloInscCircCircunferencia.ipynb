## **Semelhança entre Triângulos**
----

Se dois triângulos possuem três ângulos correspondentes congruentes e três lados correspondentes possuem a mesma razão de proporcionalidade, então dizemos que os mesmos são **semelhantes**. 

\\

### **Caso 1: Ângulo Ângulo (AA)**

\\

Se dois triângulos possuírem dois ângulos correspondentes congruentes, então eles são semelhantes.

<img src ="https://drive.google.com/uc?id=1hUtls4Ww1XQu_GTir8jPqXe4AkojDF8u"
alt="Drawing" width="300"/>


Image by <a href="https://brasilescola.uol.com.br/matematica/semelhanca-triangulos.htm">Luiz Paulo Moreira, Brasil Escola.</a>

\\






### **Caso 2: Lado Lado Lado (LLL)**

\\

Dois triângulos que possuem os três lados correspondentes proporcionais são semelhantes. 

<img src ="https://drive.google.com/uc?id=19aSKxFdGJ8b4V1F6q7VIHsDJyxJblj8m"
alt="Drawing" width="300"/>


Image by <a href="https://brasilescola.uol.com.br/matematica/semelhanca-triangulos.htm">Luiz Paulo Moreira, Brasil Escola.</a>

\\

Percebamos que as razões entre lados correspondentes são iguais:

\\

$$
\frac{AB}{DE} = \frac{BC}{EF} = \frac{CA}{FD} = \frac{1}{2}
$$

\\


### **Caso 3: Lado Ângulo Lado (LAL)**

\\

Se dois triângulos são tais que dois lados são proporcionais e o ângulo entre eles é congruente, então eles são semelhantes. 

<img src ="https://drive.google.com/uc?id=1WssfiAe9AWlr1HvKZFkXweRzIM_CniPn"

\\

Percebamos que:

\\

$$
\frac{AB}{DE} = \frac{CA}{FD} = \frac{1}{2}
$$

\\


## **Triângulos Inscritos em Circunferência**
----

Um triângulo está **inscrito** em uma circunferência se os seus vértices são também pontos da circunferência. Nesse caso, percebamos que os lados do triângulo são também cordas da circunferência. Também, podemos dizer que a circunferência está **circunscrita** ao triângulo.

\\

<img src ="https://drive.google.com/uc?id=1jRx63kkxhX4tfb1GDF51iSB-xbobpCzi"
alt="Drawing" width="300"/>


\\







Podemos calcular a área de um triângulo (acutângulo, retângulo, obtusângulo) inscrito em uma circunferência em função do comprimento de seus lados e do raio $r$ da circunferência. Seja a figura abaixo.

\\

<img src ="https://drive.google.com/uc?id=17KJHa9jCTWejKnOSX45NU3GItwcZZm3r"
alt="Drawing" width="300"/>

\\

Notemos que $AD$ é a altura do triângulo $ABC$ em relação à base $BC$. Assim, temos o triângulo retângulo $ADB$. Com o diâmetro da circunferência, $AE$, obtemos outro triângulo retângulo: $ACE$. Mas, os ângulos inscritos $A\hat{B}C$ e $A\hat{E}C$ enxergam o mesmo arco $\overset{\Large\frown}{CA}$. Logo, $m(A\hat{B}C) = m(A\hat{E}C) $. Então pelo Caso 1 (Ângulo Ângulo), os triângulos $ADB$ e $ACE$ são semelhantes. Desse modo:

\\

$$
\frac{AB}{AE} = \frac{AD}{AC} \Rightarrow AD = \frac{AB \times AC}{AE} \Rightarrow AD = \frac{AB \times AC}{2r} 
$$

\\

Logo, a área do triângulo $ABC$, $\Gamma_{ABC}$, é dada por:

\\

$$
\Gamma_{ABC} = \frac{BC \times AD}{2} = \frac{AB \times BC \times AC}{4r}
$$

## **Exercício 1: Resolvido**
----

Um triângulo retângulo está inscrito em uma circunferência de raio $r$. Os catetos possuem medida de AB e BC cm. Calcule o valor do raio $r$ da circunferência circunscrita ao triângulo.

\\


<img src ="https://drive.google.com/uc?id=1qYEFPHEwCbnSJWsU4jsNo4309EPAbJVJ"
alt="Drawing" width="300"/>


\\

*Resposta*: Para resolver esse exercício, devemos:

\\

1.   Calcular o comprimento da hipotenusa (AC);
2.   Chamar a função ``identificar_triangulo_lado`` para saber se o triângulo existe e, caso sim, saber qual o tipo do triângulo (em termos de lado);
3.   Chamar a função ``calcular_area_triangulo`` (cálculo da area);
4.   Calcular o raio $r$ por meio da fórmula de calcular área do triângulo inscrito ($\Gamma_{ABC}$ acima).

\\




```
def calcular_raio_triangulo_retangulo_inscrito(ab, bc):
  raio = sys.float_info.max
  # Os lados devem ser maior do que Zero
  if ab<=0 or bc<=0:
    print("A medida dos lados do triângulo deve ser maior do que Zero!")  
  else:
    ac = math.sqrt(ab**2 + bc**2) # Calculando hipotenusa
    t = identificar_triangulo_lado(ab, bc, ac)
    if (t != 'nao'):
      area = calcular_area_triangulo(ab, bc, ac, t)
      print('Area do Triângulo:', area)
      raio = (ab*bc*ac)/(4*area) # Da fórmula acima de cálculo de área de triângulo inscrito

  return Fraction(raio).limit_denominator()
```

```
raio_tri = calcular_raio_triangulo_retangulo_inscrito(6, 8)
print('Raio da Circunferência Circunscrita ao Triângulo: {}'.format(raio_tri))
print('#'*40+'\n')
```

## **Triângulos Circunscritos à Circunferência**
----

Considere a figura a seguir em que mostramos um triângulo **circunscrito** à circunferência. Podemos também dizer que a circunferência está **inscrita** no triângulo. Percebamos que os pontos $P$, $Q$ e $R$ são tangentes à circunferência de centro $O$ e raio $r$.

\\


<img src="https://drive.google.com/uc?id=188flDhEUiDYfuG59eXwVpOFfEMxVhRcw" alt="Drawing" width="450"/>


\\

Também podemos calcular a área de um triângulo circunscrito a uma circunferência em função do comprimento de seus lados e do raio $r$ da circunferência. Notemos que a área do triângulo $ABC$ é a soma das áreas dos triângulos $ABO$, $ACO$ e $BCO$. E que uma altura que pode ser considerada para cada um desses triângulos é o raio $r$ da circunferência inscrita ao triângulo. Logo:

$$
\Gamma_{ABC} = \Gamma_{ABO} + \Gamma_{ACO} + \Gamma_{BCO}  \Rightarrow \Gamma_{ABC} = \frac{AB \times r}{2} +  \frac{AC \times r}{2} +  \frac{BC \times r}{2}
$$

\\

Então concluímos, considerando que $p$ abaixo é o semiperímeto do triângulo:

$$
\Gamma_{ABC} = r \times \bigg ( \frac{AB + AC + BC}{2} \bigg ) \Rightarrow \Gamma_{ABC} = r \times p
$$

\\




## **Exercício 2: Resolvido**
----

Um triângulo retângulo está circunscrito a uma circunferência de raio $r$. Os catetos possuem medida de AB e BC cm. Calcule o valor do raio $r$ da circunferência inscrita ao triângulo.

\\

*Resposta*: Para resolver esse exercício, devemos:

\\

1.   Calcular o comprimento da hipotenusa (AC);
2.   Chamar a função ``identificar_triangulo_lado`` para saber se o triângulo existe e, caso sim, saber qual o tipo do triângulo (em termos de lado);
3.   Chamar a função ``calcular_area_triangulo`` (cálculo da area);
4.   Calcular o raio $r$ por meio da fórmula de calcular área do triângulo circunscrito ($\Gamma_{ABC}$ acima).

```
def calcular_raio_triangulo_retangulo_circunscrito(ab, bc):
  raio = sys.float_info.max
  # Os lados devem ser maior do que Zero
  if ab<=0 or bc<=0:
    print("A medida dos lados do triângulo deve ser maior do que Zero!")  
  else:
    ac = math.sqrt(ab**2 + bc**2) # Calculando hipotenusa
    t = identificar_triangulo_lado(ab, bc, ac)
    if (t != 'nao'):
      area = calcular_area_triangulo(ab, bc, ac, t)
      print('Area do Triângulo:', area)
      semiperimetro = (ab+bc+ac)/2
      print('Semiperímetro do Triângulo:', semiperimetro)
      raio = area/semiperimetro # Da fórmula acima de cálculo de área de triângulo circunscrito

  return Fraction(raio).limit_denominator()
  
```

```
raio_tri = calcular_raio_triangulo_retangulo_circunscrito(6, 8)
print('Raio da Circunferência Inscrita ao Triângulo: {}'.format(raio_tri))
print('#'*40+'\n')
```
