## **Regras de Precedência**

As regras de precedência determinam a ordem de avaliação das operações aritméticas. Python segue a mesma convenção usada na matemática. Em outra palavras, a ordem de avaliação dos operadores, do maior para o de menor precedência é:


*   parênteses;
*   exponenciação (potenciação);
*   multiplicação e divisão, que possuem a mesma precedência;
*   adição e subtração, que possuem a mesma precedência.

\\

<img src="https://drive.google.com/uc?id=1IRWjVQ04Hc1SBFi0e0XycRCySo9tlUga" alt="Drawing" width="600"/>

\\

A ordem de avaliação dos operadores com a mesma precedência é da esquerda para a direita. Uma exceção a essa regra é a exponenciação (potenciação).

\\

*Importante:* Na dúvida, ou mesmo que seja aparentemente óbvio, tente sempre usar parênteses nas suas expressões matemáticas. Assim, o código fica mais legível e você evita introduzir defeitos no seu código.




```
x = 2 + 3 * 2
print("Precedência 1: ", x)
```

```
x = 10 * 3 / 2 + 5 
print("Precedência 2: ", x)

x = 10 * (3 / 2 + 5)
print("Precedência 3: ", x)
```

```
x = 2 ** 2 ** 3
print("Precedência 4: ", x)

x = (2 ** 2) ** 3
print("Precedência 5: ", x)
```

```
x = 5 + 10 * 3 ** 2 * 5
print("Precedência 6: ", x)

x = (5 + 10) * 3 ** 2 * 5
print("Precedência 7: ", x)
```
